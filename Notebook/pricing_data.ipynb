```python
import pandas as pd
from pathlib import Path
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns

%matplotlib inline
```

```python
pricing = Path("c:/Users/amirh/Project-1/Data/Raw/prices_in_new_york.csv")
```

```python
pricing = pd.read_csv(pricing)
```

```python
pricing.head()
```

```python
mask = pricing.City.str.count(',') > 1
pricing[mask].head()
```

```python
us_pricing = pricing[pricing.City.str.count(',') > 1]
us_pricing.head()
```

```python
us_pricing
```

```python
us_pricing[['City Name', 'State', 'Country']] = us_pricing.City.str.split(',', expand=True)
```

```python
us_pricing
```

```python
monthly_income=us_pricing[['City Name', 'State','Average Monthly Net Salary']]
```

```python
monthly_income.dtypes
```

```python
monthly_income['Average Monthly Net Salary'] = monthly_income['Average Monthly Net Salary'].str.replace(',', '').astype(float).astype(int)
```

```python
monthly_income_sorted= monthly_income.sort_values(by='Average Monthly Net Salary')
```

```python
monthly_income_sorted
```

```python
monthly_income_sorted['Average Annual Net Salary']= monthly_income_sorted['Average Monthly Net Salary']*12
```

```python
monthly_income_sorted.plot(x='City Name', y='Average Monthly Net Salary',kind='bar', rot=45, figsize=(20, 10))
```

```python
monthly_income_sorted.plot(x='City Name', y='Average Annual Net Salary',kind='bar', rot=45, figsize=(20, 10))
```
