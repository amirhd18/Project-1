```python
import pandas as pd
from pathlib import Path
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bank_path = Path("c:/Users/amirh/Project-1/Data/Raw/Bank_Customer_retirement.csv")
```

```python
retirement_data = pd.read_csv(bank_path)
```

```python
retirement_data.head()
```

```python
# After importing the data set as a CSV file, I went through and;
# -removed two columns (customer ID and Retire)
# -converted the numbers to integers
# -Then continued forward to create visualizations with that data.
```

```python
retirement_data=retirement_data.drop(columns=["Customer ID","Retire"])
```

```python
retirement_data.head()
```

```python
retirement_data= retirement_data.astype(int)
```

```python
retirement_data.head()
```

```python
fig1=retirement_data.groupby('Age')['401K Savings'].mean().plot(title='Average Account Balance',figsize=(12,6))
```

```python
fig1.figure.savefig("Average_Account_Balance.png")
```

```python
retirement_data.columns
```

```python
retirement_data.head()
```

```python
bins = [20, 29, 39, 49, 67, 100]
labels =["20's", "30's", "40's",'50-67', '67+']
```

```python
retirement_data['age-group'] = pd.cut(retirement_data['Age'], bins=bins, labels=labels)
```

```python
retirement_data
```

```python
retirement_data.groupby('age-group')['401K Savings'].mean().plot.bar(title='Average 401K Balance', rot=45, figsize=(12,6))
```

```python
retirement_data[retirement_data['age-group'] == "30's"].plot(kind='bar', x="Age", figsize=(30,15))
```

```python
retirement_data_value_count=retirement_data['401K Savings'].value_counts(bins=5)
```

```python
retirement_data_value_count
```

```python
retirement_data_value_count.plot(kind='bar', rot=45)
```
