```python
mask = living_cost.City.str.count(',') > 1
living_cost[mask].head()
```

```python
united_states = living_cost[living_cost.City.str.count(',') > 1]
```

```python
united_states[['City Name', 'State', 'Country']] = united_states.City.str.split(',', expand=True)
```

```python
from ipywidgets import interact
import panel as pn
```

```python
united_states['State'] = united_states['State'].str.lstrip()
```

```python
#united_states = united_states.reset_index()
#united_states.drop(columns=['index', 'level_0'], inplace=True)
united_states
```

```python
list_of_states = sorted(united_states.State.unique().tolist())
```

```python
def cost_comparison(state):
    df = united_states[united_states['State'] == state]
    chart = df.hvplot(x='City Name', 
                     y='Cost of Living Index', 
                     kind='bar', rot=45).opts(fontscale=1, title='City Cost Comparison')
    chart_1 = df.hvplot(x='City Name', 
                     y='Rent Index', 
                     kind='bar').opts(fontscale=1, title='Rent Comparison')
    chart_2 = df.hvplot(x='City Name', 
                     y='Groceries Index', 
                     kind='bar', rot=45).opts(fontscale=1, title='Groceries Comparison')
    chart_3 = df.hvplot(x='City Name', 
                     y='Local Purchasing Power Index', 
                     kind='bar', rot=45).opts(fontscale=1, title='Purchasing Power Comparison')
    
    return pn.Column(f'# Analysis for {state}', pn.Row(chart, chart_1), pn.Row(chart_2, chart_3))
```

```python
interact(cost_comparison, state=list_of_states)
```
